## Estatistica dos Dados

### Variaveis Qualitativas

```python
df[['sex', 'smoker', 'day', 'time']].mode()
```

### Variaveis Quantitativas

```python
quantitative_vars = ['total_bill','tip','size']

media = df[quantitative_vars].mean()
mediana = df[quantitative_vars].min()
desvio_padrao = df[quantitative_vars].std()

print("MÃ©dia:")
print(media)
print("")

print("Mediana:")
print(mediana)
print("")

print("Desvio PadrÃ£o:")
print(desvio_padrao)
```

```python
df.describe()
```

### Calculo Estatistico para cada variavel categÃ³rica

```python
qualitative_vars = ['sex', 'smoker', 'day', 'time']

for variavel in qualitative_vars:
    valores_unicos = df[variavel].unique()
    for valor in valores_unicos:
        print(f"{variavel} - {valor}")
        subamostra = df[df[variavel] == valor]
        display(subamostra.describe())
        print("="*100)
```
